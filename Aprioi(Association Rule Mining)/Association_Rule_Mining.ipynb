```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=df.fillna('')
```

```python
df.isnull().sum()
```

```python
oht=pd.get_dummies(df)
```

```python
oht
```

```python
frequent_item_sets=apriori(oht,min_support=0.4,use_colnames=True)
```

```python
frequent_item_sets.head(5)
```

```python
rules=association_rules(frequent_item_sets,min_threshold=0.7,metric='confidence')
```

```python
rules.head(5)
```

```python
rules['antecedents_len']=rules['antecedents'].apply(lambda x:len(x))
```

```python
rules=rules[rules['antecedents_len']>1]
```

```python
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
```
