```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('/content/drive/MyDrive/DataScience-Datasets/mpg_raw (1).csv')
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].mean())
```

```python
df.isnull().sum()
```

```python
plt.hist(df['mpg'],bins=10,color='pink',edgecolor='black')
plt.xlabel('MPG')
plt.title('MPG - Histogram')
plt.show()
```

```python
sns.violinplot(x=df['cylinders'],y=df['weight'])
plt.xlabel('Cylinders')
plt.ylabel('Weight')
plt.title('Cylinders Vs Weight - Violin Plot')
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.title('MPG')
plt.show()
```

```python
q1=df['mpg'].quantile(0.25)
q3=df['mpg'].quantile(0.75)
iqr=q3-q1
lf=q1-1.5*iqr
uf=q3+1.5*iqr
df=df[df['mpg']<uf]
```

```python
plt.boxplot(df['mpg'])
plt.title('MPG')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt='g')
plt.figure(figsize=(25,7))
plt.show()
```

```python
df.describe()
```

```python
minmaxscaler=MinMaxScaler()
x=df.select_dtypes(exclude=['object']).columns
columns_to_scale=x
df[columns_to_scale]=minmaxscaler.fit_transform(df[columns_to_scale])
```

```python
df.describe()
```

```python
sns.scatterplot(x=df['mpg'],y=df['weight'])
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG Vs Weight - Scatter Plot')
plt.show()
```

```python
sns.countplot(df['weight'])
plt.title('Weight - Count Plot')
plt.show()
```

```python
sns.pointplot(x='mpg',y='weight',hue='cylinders',data=df)
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG Vs Weight - Point Plot')
plt.show()
```

```python
sns.barplot(x=df['mpg'],y=df['weight'])
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG Vs Weight - Bar Plot')
plt.show()
```

```python
sns.pairplot(df)
```
